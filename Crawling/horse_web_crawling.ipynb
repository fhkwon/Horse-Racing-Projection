```python
horse_list = []
for j in df["마번"]:
    url = f"https://studbook.kra.co.kr/html/info/ind/s_race_result.jsp?mabun=00{str(j)}"
    horse_url=requests.get(url, headers = headers, verify = False)
    soup=BeautifulSoup(horse_url.content,"html.parser")
    num = len(soup.find_all("tbody")[0].find_all("tr"))
    for i in range(num):
        A = soup.find_all("tr")[0].find("td").text
        B = soup.find_all("tr")[1].find_all("td")[1].text
        C = soup.find_all("tr")[1].find_all("td")[2].text
        D = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[0].text
        E = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[1].text
        F = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[2].text
        G = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[3].text
        H = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[4].text
        I = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[5].text
        J = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[6].text
        K = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[8].text
        L = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[9].text
        N = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[10].text
        M = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[11].text
        O = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[12].text
        P = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[13].text
        Q = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[14].text
        R = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[15].text
        S = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[16].text
        T = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[17].text
        U = soup.find_all("tbody")[0].find_all("tr")[i].find_all("td")[18].text
        horse_list.append([A, B, C, D, E, F, G, H, I, J,K, L, N, M, O, P, Q, R, S, T, U])
    print(f"현재 번호 : {j}", A)
horse_list    
```

```python
horse_dict = pd.DataFrame(horse_list, columns = ["마명", "승률" , "복승률", "경주일", "경주", "경주명", "거리", "군(등급)", "번호", 
"순위", "출전두수", "주로상태", "기록", "S1F기록", "G3F기록", "G1F기록", "수득상금", "부담중량", "마체중", "기수", "조교사"])
```
